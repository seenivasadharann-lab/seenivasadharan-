Time Series Analysis

Views Trend

```python
df.groupby(
    'Video publish time'
)['Views'].sum().plot(
    figsize=(12,5)
)

plt.title("Views Trend")
plt.show()
```

Watch Time Trend

```python
df.groupby(
    'Video publish time'
)['Watch time (hours)'].sum().plot(
    figsize=(12,5)
)

plt.title("Watch Time Trend")
plt.show()
```

Correlation Analysis

```python
corr = df[
    ['Views',
     'Watch time (hours)',
     'Subscribers']
].corr()

corr
```

Correlation Heatmap

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

Visualizations

```python
Top 10 Videos by Views
```

```python
plt.figure(figsize=(10,5))

sns.barplot(
    x='Views',
    y='Video title',
    data=top_views.head(10)
)

plt.title("Top 10 Videos by Views")
plt.show()
```

2. Watch Time Distribution

```python
plt.figure(figsize=(4,2))

sns.histplot(
    df['Watch time (hours)'],
    bins=20
)

plt.title("Watch Time Distribution")
plt.show()
```

3. Views vs Subscribers

```python
plt.figure(figsize=(5,3))

sns.scatterplot(
    x='Views',
    y='Subscribers',
    data=df
)

plt.title("Views vs Subscribers")
plt.show()
```

5. Monthly Views Trend

```python
df['Video publish time'] = pd.to_datetime(
    df['Video publish time']
)

monthly_views = df.groupby(
    df['Video publish time'].dt.month
)['Views'].sum()

plt.figure(figsize=(10,5))

monthly_views.plot(
    marker='o'
)

plt.title("Monthly Views Trend")
plt.xlabel("Month")
plt.ylabel("Views")
plt.show()
```

6. Top 5 Videos Share of Views

```python
top5 = df.nlargest(5, 'Views')

plt.figure(figsize=(7,7))

plt.pie(
    top5['Views'],
    labels=top5['Video title'],
    autopct='%1.1f%%'
)

plt.title("Top 5 Videos by Views")
plt.show()
```
